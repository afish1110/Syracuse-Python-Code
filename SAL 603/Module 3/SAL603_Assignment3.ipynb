```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
from pybaseball import statcast
```

```python
##data from statcast
data = statcast(start_dt = '2019-10-30', end_dt = '2019-10-30')
pitches = data.pitch_type
pitchers = data.player_name
```

```python
##unique pitches used to break into fastballs and offspeed
pitch_arsenal = np.unique(pitches)
#unique pitchers will loop through for percentages
pitcher_use = np.unique(pitchers)
fastballs = ['FF', 'SI', 'FC']
offspeed = ['SL', 'CH', 'CU', 'ST', 'EP']
data[np.in1d(data.player_name, 'Urquidy, José')]
```

```python
for pitcher in pitcher_use:
    ##makes new dataframe with each pitcher individually
    pitcher_data = data[np.in1d(data.player_name, pitcher)]
    ##number of fasbtalls thrown
    pitcher_fastballs = np.count_nonzero(np.in1d(pitcher_data.pitch_type, fastballs))
    ##number of offspeed thrown
    pitcher_offspeed = np.count_nonzero(np.in1d(pitcher_data.pitch_type, offspeed))
    ##fastball and offspeed percentages
    fastball_percent = (pitcher_fastballs / (pitcher_fastballs + pitcher_offspeed)) * 100
    offspeed_percent = (pitcher_offspeed / (pitcher_fastballs + pitcher_offspeed)) * 100
    ##print everything in correct format
    print(str(pitcher) + ' had ' + str(offspeed_percent) + '% off-speed pitches and ' + str(fastball_percent) + '% fastball pitches.')
```
