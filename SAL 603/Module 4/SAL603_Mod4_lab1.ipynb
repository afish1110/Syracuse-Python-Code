```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
```

```python
red_sox_roster = pd.read_csv('RedSox40man2024.csv', parse_dates= ['DoB'])
##Creates a column for the age of player on opening day 'Y' gave an error with date time so used day and then divided by 365
red_sox_roster['Age-Opening-Day-2025'] = ((np.datetime64('2025-03-27') - red_sox_roster.DoB) / np.timedelta64(1, 'D')) / 365
red_sox_roster
##writes to new csv no row index and fills na as NULL
red_sox_roster.to_csv('RedSox40man2024ODAge.csv', index=False, na_rep = 'NULL')
```
