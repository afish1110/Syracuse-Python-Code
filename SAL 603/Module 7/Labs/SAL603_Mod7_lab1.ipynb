```python
##displays all outputs of each cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import pandas as pd
```

```python
##read files
ALperformance = pd.read_csv('data/teamPerformanceAL.csv')
NLperformance = pd.read_csv('data/teamPerformanceNL.csv')
MLBattendance = pd.read_csv('data/MLBAttendanceData.csv')
stad_roof = pd.read_csv('data/coveredStadiums.csv')
MLBattendance
stad_roof
```

```python
##joining AL and NL performance
MLBperfomance = pd.concat([ALperformance, NLperformance])
MLBperfomance
```

```python
##joinging MLB performance and attendance
MLBperfomanceAttendance = pd.merge(MLBattendance, MLBperfomance, left_on = ['Tm', 'Year'], right_on = ['franchID', 'yearID'])
MLBperfomanceAttendance
```

```python
##adding the stadium cover data
MLBperfomanceAttendanceStadium = pd.merge(MLBperfomanceAttendance, stad_roof, how = 'left')
##changing na in rooftype to open
MLBperfomanceAttendanceStadium.loc[MLBperfomanceAttendanceStadium['RoofType'].isna(), 'RoofType'] = 'Open'
##covered to boolean type using a temp coloumn
MLBperfomanceAttendanceStadium['TempCovered'] = MLBperfomanceAttendanceStadium['Covered'].astype(bool)
MLBperfomanceAttendanceStadium.loc[MLBperfomanceAttendanceStadium['Covered'].isna(), 'TempCovered'] = False
MLBperfomanceAttendanceStadium.loc[MLBperfomanceAttendanceStadium['Covered'].notna(), 'TempCovered'] = True
##replacing covered with the temp data and removing temp column
MLBperfomanceAttendanceStadium['Covered'] = MLBperfomanceAttendanceStadium['TempCovered']
MLBperfomanceAttendanceStadium = MLBperfomanceAttendanceStadium.drop(columns = 'TempCovered')
MLBperfomanceAttendanceStadium
```

```python
##write to a csv
MLBperfomanceAttendanceStadium.to_csv('teamPerformanceAttendance.csv')
```

```python
##aggregating attendance by mean
aggregated_attend = MLBperfomanceAttendanceStadium.groupby('Tm', as_index = False)['Attendance'].mean()
##write to csv
aggregated_attend.to_csv('teamAttendanceAverage.csv')
```
